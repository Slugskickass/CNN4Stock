Lets build a CNN to study the stock market

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

Our test stock

```python
stock = 'N91.L' # a london / Cape town based finance company
timePeriod = "3mo"
```

```python
stockTicker = yf.Ticker(stock)
stockHist = stockTicker.history(period=timePeriod)
```

```python
plt.plot(stockHist['Open'])
print(stockHist.columns())
```

Need a function to build our tensor and character:
5 by ten for the tensor and 0 or 1 for the character

```python
data = []
outcome = []
for currentPos in range(0,len(stockHist)-11, 10):
    smallData = stockHist[currentPos:currentPos+11]
    tensordata = smallData[0:10].to_numpy()[:,0:5]
    lastValue = smallData['Close'][9]
    openValue = smallData['Open'][10]
    if openValue > lastValue:
        character = 1
    else:
        character = 0
    data.append(tensordata)
    outcome.append(character)
```
